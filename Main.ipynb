## Clean and visualize data

```python
# Replace Yes or No values with 1 and 0 for binary classification
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x == 'Y' else 0)
employee_df
```

```python
# See if there is any missing data
sns.heatmap(employee_df.isnull(), yticklabels=False, cbar=False, cmap='Blues')
```

```python
# Analyze distribution of data
employee_df.hist(bins=30, figsize=(20, 20), color='r')
```

```python
# Drop EmployeeCount, StandardHours, and Over18 in memory due to low cardinality
# employee_df.drop(['EmployeeCount', 'StandardHours', 'Over18'], axis=1, inplace=True)
employee_df
```

```python
# Find the count and % of employees that left the company
left_df = employee_df[employee_df['Attrition'] == 1]
remaining_df = employee_df[employee_df['Attrition'] == 0]
print('Count of employees who left: ', left_df['Attrition'].count())
print('Count of employees who remain:', remaining_df['Attrition'].count())
print('Total count: ', employee_df['Attrition'].count())
print(f'% of employees who left: {(left_df["Attrition"].count() / employee_df["Attrition"].count()) * 100:.2f}%')
```

```python
'''
Possible factors to attrition:

AverageAge is lower for employees who left
DailyRate is lower for employees who left
DistanceFromHome is higher for employees who left
EnvironmentSatisfaction is lower for employees who left
StockOptionLevel is lower for employees who left

'''
left_df.describe()
```

## Correlation Analysis

```python
''' Top 3 Strongest Positive and Negative Correlations to Attrition:
        Positive:
            1). Overtime (weak, 0.25) - Suggests that the more overtime an employee works, the higher attrition level
            2). DistanceFromHome (weak, 0.08) - Suggests that the more distance an employee is from their home, the higher attrition level
            3). NumCompaniesWorked (weak, 0.04) - Suggests that the more companies an employee works for, the higher attrition level
        Negative:
            1). JobLevel (weak, -0.17) - Suggests that the higher the employee's job level, the lower attrition level
            2). TotalWorkingYears (weak, -0.17) - Suggests that the higher the employee's total working years, the lower attrition level
            3). Age (weak, -0.16) - Suggests that the older the employee, the lower attrition level

    Analysis: Although there is some correlation between the above factors, the overall correlation is not strong enough to attribute any individual
              factor to attrition without further investigation.
'''

# Get correlations of numeric columns only to test Attrition correlation
numeric_cols = employee_df.select_dtypes(include=['int64', 'float64']).columns
correlations = employee_df[numeric_cols].corr()
f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, ax=ax)
```
